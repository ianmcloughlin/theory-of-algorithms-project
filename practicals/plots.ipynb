```python
# Numerical arrays.
import numpy as np

# Plots.
import matplotlib.pyplot as plt
```

## $f(x) = x^2 + 1$

***

```python
# Range of x values.
x = np.linspace(-4.0, 4.0, 1000)

# Corresponding f(x) values.
f = x**2 + 1
```

```python
# Create a plot.
fig, ax = plt.subplots(figsize=(10, 6))

# Plot f(x).
ax.plot(x, f, label='$x^2+1$')

# Show the legend.
ax.legend();
```

## $g(x) = 3x^6 + x^4 +1$

***

```python
# Range of x values.
x = np.linspace(-4.0, 4.0, 1000)

# Corresponding g(x) values.
g = 3 * x**6 + x**4 + 1
```

```python
# Create a plot.
fig, ax = plt.subplots(figsize=(10, 6))

# Plot g(x).
ax.plot(x, g, label='$3x^6 + x^4 +1$')

# Show the legend.
ax.legend();
```

## $h(x) = x^2$ and $j(x) = 2^x$

***

```python
# Range of x values.
x = np.linspace(0.0, 100.0, 1000)

# Corresponding h(x) values.
h = x**2

# Corresponding j(x) values.
j = 2**x
```

```python
# Create a plot.
fig, ax = plt.subplots(figsize=(10, 6))

# Plot h(x).
ax.plot(x, h, label='$x^2$')

# Plot j(x).
ax.plot(x, j, label='$2^x$')

# Show the legend.
ax.legend();
```

## $g(f(x))$

***

```python
# Range of x values.
x = np.linspace(0.0, 100.0, 1000)

# Corresponding f(x) values.
f = x**2 + 1
```

```python
# Corresponding g(f(x)) values.
gof = 3 * f**6 + f**4 + 1
```

```python
# Create a plot.
fig, ax = plt.subplots(figsize=(10, 6))

# Plot f(x).
ax.plot(x, f, label='$f(x) = x^2+1$')

# Plot g(x).
ax.plot(x, g, label='$g(x) = 3x^6 + x^4 +1$')

# Plot g(f(x)).
ax.plot(x, gof, label='$g(f(x)) = 3(x^2 + 1)^6 + (x^2 + 1)^4 +1$')

for i in range(1, 13):
  # Plot x^i.
  ax.plot(x, x**i, label=f'$x^{i}$')

# Plot of 2^x.
ax.plot(x, 2**x, label='$2^x$')

# Show the legend.
ax.legend();
```
